```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import os
import time
from kerastuner import Hyperband
from kerastuner.engine.hyperparameters import HyperParameters

# ==================== DATA LOADING AND PREPARATION ====================
print("Loading CIFAR-10 dataset...")
# Load dataset CIFAR-10
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# Normalisasi data
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Konversi label ke one-hot encoding
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# Class
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# ==================== VISUALISASI DATA ====================
def visualize_data():
    plt.figure(figsize=(10, 10))
    for i in range(25):
        plt.subplot(5, 5, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(x_train[i])
        plt.xlabel(class_names[np.argmax(y_train[i])])
    plt.tight_layout()
    plt.savefig('cifar10_examples.png')
    plt.show()

# ==================== MODEL HYPERPARAMETER TUNING ====================
def build_model(hp):
    model_type = hp.Choice('model_type', ['custom', 'transfer_learning'])

    if model_type == 'custom':
        model = keras.Sequential()

        # Input layer
        model.add(keras.layers.InputLayer(input_shape=(32, 32, 3)))

        # Convolutional layers dengan hyperparameter tuning
        for i in range(hp.Int('num_conv_layers', 2, 4)):
            model.add(keras.layers.Conv2D(
                filters=hp.Int(f'filters_{i}', 32, 128, step=32),
                kernel_size=hp.Choice(f'kernel_{i}', [3, 5]),
                activation='relu',
                padding='same'))
            model.add(keras.layers.BatchNormalization())
            model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
            model.add(keras.layers.Dropout(
                hp.Float(f'dropout_{i}', 0.0, 0.5, step=0.1)))

        model.add(keras.layers.Flatten())

    else:
        base_model_name = hp.Choice('base_model', ['VGG16', 'MobileNetV2', 'ResNet50'])

        # Load pre-trained model with appropriate input shape for CIFAR-10
        if base_model_name == 'VGG16':
            # VGG16 expects minimum input size of 32x32
            base_model = keras.applications.VGG16(
                weights='imagenet',
                include_top=False,
                input_shape=(32, 32, 3))
        elif base_model_name == 'MobileNetV2':
            # MobileNetV2 expects minimum input size of 32x32
            base_model = keras.applications.MobileNetV2(
                weights='imagenet',
                include_top=False,
                input_shape=(32, 32, 3))
        elif base_model_name == 'ResNet50':
            # ResNet50 expects minimum input size of 32x32
            base_model = keras.applications.ResNet50(
                weights='imagenet',
                include_top=False,
                input_shape=(32, 32, 3))

        # Freeze base model layers
        freeze_percentage = hp.Float('freeze_percentage', 0.5, 1.0, step=0.1)
        num_layers = len(base_model.layers)
        num_frozen = int(freeze_percentage * num_layers)

        for layer in base_model.layers[:num_frozen]:
            layer.trainable = False

        # Create model
        model = keras.Sequential()

        # Add preprocessing specific to the chosen model
        if base_model_name == 'VGG16':
            model.add(keras.layers.Lambda(keras.applications.vgg16.preprocess_input, input_shape=(32, 32, 3)))
        elif base_model_name == 'MobileNetV2':
            model.add(keras.layers.Lambda(keras.applications.mobilenet_v2.preprocess_input, input_shape=(32, 32, 3)))
        elif base_model_name == 'ResNet50':
            model.add(keras.layers.Lambda(keras.applications.resnet50.preprocess_input, input_shape=(32, 32, 3)))

        # Add base model
        model.add(base_model)

        # Add Global Average Pooling
        model.add(keras.layers.GlobalAveragePooling2D())

    # Dense layers (common for both custom and transfer learning)
    for i in range(hp.Int('num_dense_layers', 1, 2)):
        model.add(keras.layers.Dense(
            units=hp.Int(f'units_{i}', 64, 256, step=64),
            activation='relu'))
        model.add(keras.layers.BatchNormalization())
        model.add(keras.layers.Dropout(
            hp.Float(f'dense_dropout_{i}', 0.0, 0.5, step=0.1)))

    # Output layer
    model.add(keras.layers.Dense(10, activation='softmax'))

    # Compile model
    model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Float('learning_rate', 1e-4, 1e-2, sampling='log')),
        loss='categorical_crossentropy',
        metrics=['accuracy'])

    return model

# ==================== PLOTTING FUNCTIONS ====================
def plot_training_history(history):
    plt.figure(figsize=(12, 4))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.ylim([0, 1])
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')

    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

# ==================== MAIN EXECUTION ====================
def main():
    # Visualize example images
    print("Visualizing example images...")
    visualize_data()

    # Set up logging directory with timestamp
    timestamp = time.strftime("%Y%m%d-%H%M%S")
    tuner_dir = os.path.join('tuner_dir', timestamp)
    os.makedirs(tuner_dir, exist_ok=True)

    # Configure hyperparameters for the search
    print("Setting up hyperparameters search space...")
    hp = HyperParameters()

    # Initialize tuner
    print("Initializing Keras Tuner...")
    tuner = Hyperband(
        build_model,
        objective='val_accuracy',
        max_epochs=10,
        factor=3,
        directory=tuner_dir,
        project_name='cifar10_cnn')

    # Early stopping callback
    early_stopping = keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True)

    # Model checkpoint callback
    checkpoint_path = os.path.join('checkpoints', timestamp, 'model.keras')
    os.makedirs(os.path.dirname(checkpoint_path), exist_ok=True)
    checkpoint_callback = keras.callbacks.ModelCheckpoint(
        checkpoint_path,
        save_best_only=True,
        monitor='val_accuracy')

    # TensorBoard callback
    tensorboard_dir = os.path.join('logs', timestamp)
    os.makedirs(tensorboard_dir, exist_ok=True)
    tensorboard_callback = keras.callbacks.TensorBoard(
        log_dir=tensorboard_dir,
        histogram_freq=1)

    # Set up all callbacks
    callbacks = [
        early_stopping,
        checkpoint_callback,
        tensorboard_callback
    ]

    # Search for best hyperparameters
    print("Starting hyperparameter search...")
    tuner.search(
        x_train, y_train,
        epochs=50,
        validation_split=0.2,
        callbacks=[early_stopping],
        batch_size=64)

    # Get best hyperparameters
    best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
    print(f"Best hyperparameters found: {best_hps.values}")

    # Log the model architecture type
    model_type = best_hps.values.get('model_type', 'custom')
    if model_type == 'transfer_learning':
        base_model = best_hps.values.get('base_model', 'Unknown')
        freeze_pct = best_hps.values.get('freeze_percentage', 0)
        print(f"Best model uses transfer learning with {base_model} (freeze: {freeze_pct*100:.0f}%)")
    else:
        print("Best model uses custom architecture")

    # Get best model
    print("Training final model with best hyperparameters...")
    best_model = tuner.get_best_models(num_models=1)[0]

    # Train final model with best hyperparameters
    history = best_model.fit(
        x_train, y_train,
        epochs=50,
        validation_data=(x_test, y_test),
        batch_size=64,
        callbacks=callbacks)

    # Evaluate model on test set
    print("Evaluating model on test set...")
    test_loss, test_acc = best_model.evaluate(x_test, y_test, verbose=2)
    print(f"Test accuracy: {test_acc:.4f}")
    print(f"Test loss: {test_loss:.4f}")

    # Plot training history
    plot_training_history(history)

    # Save the final model
    final_model_path = f'cifar10_model_{timestamp}.keras'
    best_model.save(final_model_path)
    print(f"Model saved to {final_model_path}")

    # Show model summary
    best_model.summary()

    # Optional: Make predictions on a few test images
    predictions = best_model.predict(x_test[:10])
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(y_test[:10], axis=1)

    print("\nExample predictions:")
    for i in range(10):
        is_correct = predicted_classes[i] == true_classes[i]
        status = "✓" if is_correct else "✗"
        print(f"Image {i}: True: {class_names[true_classes[i]]}, Predicted: {class_names[predicted_classes[i]]} {status}")

    # Visualize confusion matrix
    def plot_confusion_matrix():
        # Make predictions on the entire test set
        y_pred = best_model.predict(x_test)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)

        # Compute confusion matrix
        from sklearn.metrics import confusion_matrix, classification_report
        cm = confusion_matrix(y_true_classes, y_pred_classes)

        # Plot confusion matrix
        plt.figure(figsize=(10, 8))
        plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
        plt.title('Confusion Matrix')
        plt.colorbar()
        tick_marks = np.arange(len(class_names))
        plt.xticks(tick_marks, class_names, rotation=45)
        plt.yticks(tick_marks, class_names)

        # Add text annotations
        thresh = cm.max() / 2.
        for i in range(cm.shape[0]):
            for j in range(cm.shape[1]):
                plt.text(j, i, format(cm[i, j], 'd'),
                        horizontalalignment="center",
                        color="white" if cm[i, j] > thresh else "black")

        plt.tight_layout()
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.savefig('confusion_matrix.png')
        plt.show()

        # Print classification report
        print("\nClassification Report:")
        print(classification_report(y_true_classes, y_pred_classes, target_names=class_names))

    # Plot confusion matrix
    print("\nGenerating confusion matrix...")
    try:
        plot_confusion_matrix()
    except ImportError:
        print("scikit-learn not available. Skipping confusion matrix.")

    print("\nTraining and evaluation complete!")

if __name__ == "__main__":
    main()
```

```python
%%writefile app.py
import streamlit as st

st.set_page_config(
    page_title="CIFAR-10 Classifier",
    layout="wide"
)

import tensorflow as tf
from tensorflow import keras
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import os

# Load model
@st.cache_resource
def load_model():
    model = keras.models.load_model('cifar10_model_20250503-084455.keras')
    return model

model = load_model()

# Class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Fungsi untuk evaluasi gambar
def evaluate_image(image, model, class_names):
    try:
        # Preprocess gambar
        if image.mode != 'RGB':
            image = image.convert('RGB')

        processed_img = image.resize((32, 32))
        img_array = np.array(processed_img) / 255.0

        if img_array.shape[-1] != 3:
            st.warning(f"Gambar memiliki {img_array.shape[-1]} channel, mengkonversi ke 3 channel")
            img_array = img_array[..., :3]

        img_array = np.expand_dims(img_array, axis=0)

        # Prediksi
        predictions = model.predict(img_array)
        predicted_class = class_names[np.argmax(predictions[0])]
        confidence = np.max(predictions[0]) * 100

        return {
            'original_image': image,
            'processed_image': processed_img,
            'predicted_class': predicted_class,
            'confidence': confidence,
            'probabilities': predictions[0]
        }

    except Exception as e:
        st.error(f"Error saat memproses gambar: {str(e)}")
        return None

# Tampilan Streamlit
st.title("CIFAR-10 Classifier")
st.write("Aplikasi Untuk Klasifikasi Gambar")

tab1, tab2 = st.tabs(["Upload Gambar", "Evaluasi dengan Contoh"])

with tab1:
    st.header("Upload Gambar")
    uploaded_file = st.file_uploader("Pilih gambar...", type=["jpg", "png", "jpeg"], key="uploader1")

    if uploaded_file is not None:
        image = Image.open(uploaded_file)
        result = evaluate_image(image, model, class_names)

        if result:
            col1, col2 = st.columns(2)
            with col1:
                st.image(result['original_image'], caption="Gambar Asli", use_container_width=True)
                st.image(result['processed_image'], caption="Gambar Diproses (32x32)", use_container_width=True)

            with col2:
                st.success(f"**Prediksi:** {result['predicted_class']}")
                st.info(f"**Confidence Level:** {result['confidence']:.2f}%")

                fig, ax = plt.subplots(figsize=(10, 4))
                bars = ax.bar(class_names, result['probabilities'])
                bars[class_names.index(result['predicted_class'])].set_color('green')
                plt.xticks(rotation=45)
                plt.ylabel("Probabilitas")
                plt.ylim([0, 1])
                st.pyplot(fig)

with tab2:
    st.header("Evaluasi Dengan Gambar Contoh")

    option = st.radio("Pilih sumber gambar:",
                     ["Gambar Bawaan", "Dari URL"],
                     horizontal=True)

    if option == "Gambar Bawaan":
        example_images = {
            "Kucing": "https://upload.wikimedia.org/wikipedia/commons/4/4d/Cat_November_2010-1a.jpg",
            "Anjing": "https://upload.wikimedia.org/wikipedia/commons/d/d0/German_Shepherd_-_DSC_0346_%2810096362833%29.jpg",
            "Pesawat": "https://upload.wikimedia.org/wikipedia/commons/3/3a/Cessna_172_Skyhawk_III.jpg"
        }

        selected = st.selectbox("Pilih contoh gambar:", list(example_images.keys()))

        if st.button("Evaluasi Gambar Contoh"):
            try:
                temp_file = f"temp_{selected}.jpg"
                urllib.request.urlretrieve(example_images[selected], temp_file)
                image = Image.open(temp_file)
                result = evaluate_image(image, model, class_names)

                if result:
                    st.success(f"Hasil evaluasi untuk gambar {selected}:")

                    col1, col2 = st.columns(2)
                    with col1:
                        st.image(result['original_image'], caption="Gambar Asli", use_container_width=True)
                        st.image(result['processed_image'], caption="Gambar Diproses (32x32)", use_container_width=True)

                    with col2:
                        st.metric("Prediksi", result['predicted_class'])
                        st.metric("Confidence", f"{result['confidence']:.2f}%")

                        fig, ax = plt.subplots(figsize=(10, 4))
                        bars = ax.bar(class_names, result['probabilities'])
                        bars[class_names.index(result['predicted_class'])].set_color('green')
                        plt.xticks(rotation=45)
                        plt.ylabel("Probabilitas")
                        plt.ylim([0, 1])
                        st.pyplot(fig)

                os.remove(temp_file)

            except Exception as e:
                st.error(f"Gagal memproses gambar contoh: {str(e)}")
                if os.path.exists(temp_file):
                    os.remove(temp_file)

    else:
        url = st.text_input("Masukkan URL gambar:")
        if url and st.button("Evaluasi dari URL"):
            try:
                temp_file = "temp_url_image.jpg"
                urllib.request.urlretrieve(url, temp_file)
                image = Image.open(temp_file)
                result = evaluate_image(image, model, class_names)

                if result:
                    st.success("Hasil evaluasi dari URL:")

                    col1, col2 = st.columns(2)
                    with col1:
                        st.image(result['original_image'], caption="Gambar Asli", use_container_width=True)
                        st.image(result['processed_image'], caption="Gambar Diproses (32x32)", use_container_width=True)

                    with col2:
                        st.metric("Prediksi", result['predicted_class'])
                        st.metric("Confidence", f"{result['confidence']:.2f}%")

                        fig, ax = plt.subplots(figsize=(10, 4))
                        bars = ax.bar(class_names, result['probabilities'])
                        bars[class_names.index(result['predicted_class'])].set_color('green')
                        plt.xticks(rotation=45)
                        plt.ylabel("Probabilitas")
                        plt.ylim([0, 1])
                        st.pyplot(fig)

                os.remove(temp_file)

            except Exception as e:
                st.error(f"Gagal memproses URL gambar: {str(e)}")
                if os.path.exists(temp_file):
                    os.remove(temp_file)

# Sidebar dengan info tambahan
st.sidebar.header("Informasi Model")
st.sidebar.write(f"Model CNN untuk klasifikasi 10 kelas")
st.sidebar.write("Kelas yang dikenali:")
st.sidebar.write(", ".join(class_names))

st.sidebar.header("Petunjuk")
st.sidebar.markdown("""
1. Gunakan tab **Upload Gambar** untuk menguji gambar Anda
2. Gunakan tab **Evaluasi dengan Contoh** untuk menguji dengan gambar bawaan atau URL
3. Gambar akan ditampilkan dalam ukuran asli dan versi 32x32 yang diproses
""")
```
